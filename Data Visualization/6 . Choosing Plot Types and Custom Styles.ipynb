```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

1. sns.lineplot- 꺾은선형 차트는 일정 기간 동안의 추세를 표시하는 데 가장 적합하며, 여러 선을 사용하여 둘 이상의 그룹의 추세를 표시할 수 있습니다.
2. sns.barplot- 막대 차트는 다양한 그룹에 해당하는 수량을 비교하는 데 유용합니다.
3. sns.heatmap- 히트맵을 사용하면 숫자 표에서 색상으로 구분된 패턴을 찾을 수 있습니다.
4. sns.scatterplot- 산점도는 두 개의 연속 변수 사이의 관계를 보여줍니다. 색상으로 구분된 경우 세 번째 범주형 변수 와의 관계를 표시할 수도 있습니다 .
5. sns.regplot- 산점도에 회귀선을 포함하면 두 변수 간의 선형 관계를 더 쉽게 확인할 수 있습니다.
6. sns.lmplot- 이 명령은 분산형 차트에 여러 색상으로 구분된 그룹이 포함된 경우 여러 회귀선을 그리는 데 유용합니다.
7. sns.swarmplot- 범주형 산점도는 연속형 변수와 범주형 변수 사이의 관계를 보여줍니다.
8. sns.histplot- 히스토그램은 단일 수치 변수의 분포를 보여줍니다.
9. sns.kdeplot- KDE 도표 (또는 2D KDE 도표 )는 단일 수치 변수(또는 두 수치 변수)의 추정된 매끄러운 분포를 보여줍니다.
10. sns.jointplot- 이 명령은 각 개별 변수에 해당하는 KDE 도표와 함께 2D KDE 도표를 동시에 표시하는 데 유용합니다.

```python
spotify_filepath = "./input/spotify.csv"

# Read the file into a variable spotify_data
spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)

# Line chart 
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
```
